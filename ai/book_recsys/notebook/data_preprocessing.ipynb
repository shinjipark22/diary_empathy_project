```python
input_file = "../data/raw/aladin_books_bestseller.csv" 
output_file = "../data/processed/aladin_books_bestseller_cleaned.csv"
```

```python
df = pd.read_csv(input_file)

# 1. description 결측치를 title로 보완
df['description'] = df['description'].fillna(df['title'])

# 2. 임베딩 성능을 위한 텍스트 피처 생성
def create_feature_text(row):
    title = str(row['title'])
    desc = str(row['description'])
    cat = str(row['category']).split('>')[1] if isinstance(row['category'], str) and '>' in row['category'] else ""
    return f"제목: {title} | 장르: {cat} | 소개: {desc}"

df['feature_text'] = df.apply(create_feature_text, axis=1)

# 3. 결과 저장
os.makedirs("../data", exist_ok=True)
df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f"전처리 완료. 저장 경로: {output_file}")
```
